```python
import os
import random
from pydub import AudioSegment
import librosa
import pandas as pd
import numpy as np
from tqdm import tqdm
import soundfile as sf

# Files Count
max_files = 1000

# Load the mp3 file
audio = AudioSegment.from_mp3("/Users/owo/HOUSE/@DUNE/@AI/Sound Classification/newsound/Glass shattering sound effect.mp3")

# Convert mp3 to wav
audio.export("/Users/owo/HOUSE/@DUNE/@AI/Sound Classification/newsound/Glass shattering sound effect.wav", format="wav")

# Load the wav file with librosa
y, sr = librosa.load("/Users/owo/HOUSE/@DUNE/@AI/Sound Classification/newsound/Glass shattering sound effect.wav")

# Calculate the number of frames for 2 second audio
num_frames = sr * 2

# Calculate the total number of 2-second chunks that can be made
num_chunks = len(y) // num_frames

# Generate random starting points for each chunk
start_points = random.sample(range(len(y) - num_frames), num_chunks)

# Split the audio into chunks at the random starting points
chunks = np.array([y[start:start + num_frames] for start in start_points])

# Create a new dataframe to store the information
new_data = pd.DataFrame(columns=['slice_file_name', 'fsID', 'start', 'end', 'salience', 'fold', 'classID', 'class'])

# Save each chunk to a new wav file and add the information to the dataframe
for i, (chunk, start) in tqdm(enumerate(zip(chunks, start_points))):
    if i >= max_files:
        break
    filename = f"glass_shatter_{i}.wav"
    sf.write(f"/Users/owo/HOUSE/@DUNE/@AI/Sound Classification/newsound/glass shatter/{filename}", chunk, sr)
    row = pd.DataFrame({'slice_file_name': [filename], 'fsID': ['111111'], 'start': [start/sr], 'end': [(start+num_frames)/sr], 'salience': ['1'], 'fold': [11], 'classID': ['11'], 'class': ['glass_shatter']})
    new_data = pd.concat([new_data, row], ignore_index=True)

# Save the new dataframe to a csv file
new_data.to_csv("/Users/owo/HOUSE/@DUNE/@AI/Sound Classification/newsound/UrbanSound8K_for_glassshatter.csv", index=False)
```
