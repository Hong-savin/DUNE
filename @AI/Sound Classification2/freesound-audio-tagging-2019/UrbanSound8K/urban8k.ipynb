```python
#Only need siren

import os
import shutil
import pandas as pd
metadata = pd.read_csv('UrbanSound8K.csv')

# Filter the dataframe to include only rows where the class is "siren"
siren_sounds = metadata[metadata['class'] == 'siren']

# Display the first few rows of the filtered dataframe
siren_sounds.head()


# Create a new directory for siren sounds
os.makedirs('siren', exist_ok=True)

# Iterate over all files in the dataset
for index, row in siren_sounds.iterrows():
    # Get the path to the source file
    source = os.path.join('fold' + str(row['fold']), row['slice_file_name'])
    # print(source)
    # Check if the source file exists
    if os.path.exists(source):
        # Get the path to the destination file
        destination = os.path.join('siren', row['slice_file_name'])
        
        # Copy the file to the new directory
        shutil.copy(source, destination)
        print("moved")
```

```python
#Only need car horn

metadata = pd.read_csv('UrbanSound8K.csv')

import os
import shutil

# Create a new directory for car_horn sounds
os.makedirs('car_horn', exist_ok=True)

# Filter the dataframe to include only rows where the class is "car_horn"
car_horn_sounds = metadata[metadata['class'] == 'car_horn']
count = 0
# Iterate over all files in the dataset
for index, row in car_horn_sounds.iterrows():
    # Get the path to the source file
    source = os.path.join('fold' + str(row['fold']), row['slice_file_name'])
    
    # Check if the source file exists
    if os.path.exists(source):
        # Get the path to the destination file
        destination = os.path.join('car_horn', row['slice_file_name'])
        
        # Copy the file to the new directory
        shutil.copy(source, destination)
        count = count + 1
print(count)
```

```python
# Create a new dataframe with the required format from the original metadata
car_horn_sounds_new_format_from_original = metadata[metadata['class'] == 'car_horn'][['slice_file_name', 'class']].copy()
car_horn_sounds_new_format_from_original.columns = ['fname', 'labels']

# Save the new dataframe to a CSV file
car_horn_sounds_new_format_from_original.to_csv('car_horn_sounds_new_format_from_original.csv', index=False)
```
